# `view_capture` — quick demo

Rotate a model in the interactive viewer, click **📷 Capture orientation**, and render the figure from that exact angle. The whole tool is one file: `view_capture.py`.

## 1. Get the tool

On Colab/fresh env, also install ASE (uncomment below).

```python
# On Colab or a fresh environment, uncomment:
# !pip install -q ase

# If view_capture isn't importable, set this once to your hosted raw URL:
VIEW_CAPTURE_URL = "https://raw.githubusercontent.com/MineBonik/view_capture/main/view_capture.py"
try:
    from view_capture import view_capture, CAPTURED_VIEWS, set_capture
except ImportError:
    if "<" not in VIEW_CAPTURE_URL:
        import urllib.request; urllib.request.urlretrieve(VIEW_CAPTURE_URL, "view_capture.py")
        from view_capture import view_capture, CAPTURED_VIEWS, set_capture
    else:
        raise ImportError("Set VIEW_CAPTURE_URL above, or put view_capture.py next to this notebook.")
print('view_capture ready')
```

## 2. Build a small model

```python
from ase.build import fcc111, add_adsorbate

slab = fcc111('Cu', size=(3, 3, 3), vacuum=6.0)
add_adsorbate(slab, 'O', height=1.6, position='fcc')
len(slab)
```

## 3. View & capture

Rotate the model to the perspective you want, then click **📷 Capture orientation**. The rotation string is copied to your clipboard and (on Colab) stored in `CAPTURED_VIEWS['demo']`.

```python
view_capture(slab, name='demo')
```

## 4. Render from the captured angle

PNG via ASE (no POV-Ray needed). If you didn't click the button, it falls back to a default angle so the cell still runs.

```python
from ase.io import write
from IPython.display import Image

rot = CAPTURED_VIEWS.get('demo', {}).get('rotation', '-75x,0y,0z')
write('demo.png', slab, rotation=rot, radii=0.9, show_unit_cell=2)
Image('demo.png')
```
